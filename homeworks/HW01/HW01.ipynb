```python
# Импорт numpy
import numpy as np
```

```python
# Создание массива
arr = np.array([1, 2, 3, 4, 5])
arr2 = np.array([4, 7, 1, 5, 2])
print(arr)
```

```python
arr_zeros = np.zeros((2, 3)) # создаст матрицу 2x3 из нулей
arr_ones = np.ones((3, 2)) # создаст матрицу 3x2 из единиц
arr_arange = np.arange(0, 10, 2) # аналог range
print(arr_zeros, arr_ones, arr_arange)
```

```python
arr_shape = arr_arange.shape # форма матрицы
arr_dtype = arr_arange.dtype # тип данных 
print(arr_shape, arr_dtype)
```

```python
# Элемент через индекс
single = arr[2]
print(single)
```

```python
# Срез массива
sub_arr = arr[1:4]
print(sub_arr)
```

```python
# Векторные операции
vec_sum = arr + arr2
vec_mul = arr * arr2
print(vec_sum, vec_mul)
```

```python
# Вычисление средней величины
mean_value = arr.mean()
print(mean_value)
```
